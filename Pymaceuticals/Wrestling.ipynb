```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
input1 = input("name a wrestler")
```

```python
wwe13 = pd.read_csv("Resources/WWE-Data-2013.csv")
wwe14 = pd.read_csv("Resources/WWE-Data-2014.csv")
wwe15 = pd.read_csv("Resources/WWE-Data-2015.csv")
wwe16 = pd.read_csv("Resources/WWE-Data-2016.csv")
```

```python
combined = wwe13.merge(wwe14, on ="Wrestler",how='outer').merge(wwe15, on = "Wrestler",how='outer').merge(wwe16, on = "Wrestler",how='outer')
combined.columns = ['Wrestler','Wins2013','Losses2013','Draws2013','Wins2014','Losses2014','Draws2014','Wins2015','Losses2015','Draws2015','Wins2016','Losses2016','Draws2016']
combined = combined.set_index(['Wrestler'])
combined.fillna(0,inplace = True)
combined
```

```python
temp = combined[combined.index == input1]
temp_wins = temp[['Wins2013','Wins2014','Wins2015','Wins2016']]
temp_losses = temp[['Losses2013','Losses2014','Losses2015','Losses2016']]
temp_draws = temp[['Draws2013','Draws2014','Draws2015','Draws2016']]
wins = temp_wins.loc[input1]
losses = temp_losses.loc[input1]
draws = temp_draws.loc[input1]
```

```python
years = [2013,2014,2015,2016]
x_axis = np.arange(len(years))
plt.plot(x_axis, wins, color="red")
plt.plot(x_axis, losses, color="blue")
plt.plot(x_axis, draws, color="green")
plt.show()
```

```python
plt.title("The Bloodthirst of Kings")
plt.ylabel("Number of Battles")
plt.xlabel("King")
```
